```python
import librosa
import librosa.display
import scipy
import os
import pathlib
import numpy
import sklearn
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (14, 4)
```

```python
filename = '125_bounce.wav'
y, sr = librosa.load(os.path.join("work","audio",filename))
```

```python
librosa.display.waveplot(y, sr)
```

```python
ipd.Audio(y, rate = sr)
```

```python
onset_frames = librosa.onset.onset_detect(y, sr=sr)
onset_times = librosa.frames_to_time(onset_frames, sr=sr)
onset_samples = librosa.frames_to_samples(onset_frames)
```

```python
y_with_beeps = librosa.clicks(onset_times, sr=sr, length=len(y))
ipd.Audio(y + y_with_beeps, rate=sr)
```

```python
def extract_features(y, fs):
    zcr = librosa.zero_crossings(y).sum()
    energy = scipy.linalg.norm(y)
    # equivalent to sum(abs(y**2)) when non complex values are  NOT present
    return [zcr, energy]
```

```python
frame_sz = sr*0.090
features = numpy.array([extract_features(y[i:int(i+frame_sz)], sr) for i in onset_samples])
features[0,]
```

```python
min_max_scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(-1, 1))
features_scaled = min_max_scaler.fit_transform(features)
plt.scatter(features_scaled[:,0], features_scaled[:,1])
plt.xlabel('Zero Crossing Rate (scaled)')
plt.ylabel('Spectral Centroid (scaled)')
```

```python
model = sklearn.cluster.KMeans(n_clusters=2)
labels = model.fit_predict(features_scaled)
print(labels)
```

```python
x_with_beeps = librosa.clicks(onset_times[labels==0], sr = sr, length=len(y))
ipd.Audio(y + x_with_beeps, rate=sr)
```

```python
x_with_beeps = librosa.clicks(onset_times[labels==1], sr = sr, length=len(y))
ipd.Audio(y + x_with_beeps, rate=sr)
```

```python
plt.scatter(features_scaled[labels==0,0], features_scaled[labels==0,1], c='b')
plt.scatter(features_scaled[labels==1,0], features_scaled[labels==1,1], c='r')
plt.xlabel('Zero Crossing Rate (scaled)')
plt.ylabel('Energy (scaled)')
plt.legend(('Class 0', 'Class 1'))
```
